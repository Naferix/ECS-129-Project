```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
###make sure paths is calibrated

no_knot = open(r'1timA.txt')

###

x_val = []
y_val = []
z_val = []

coords1 = []
for line in no_knot:
    tempvar = ''
    templist = []
    for letter in line:
        if letter != ' ':
            tempvar = tempvar + letter
            if letter == line[-1]:
                templist.append(tempvar.strip())
        else:
            if tempvar != '':
                templist.append(tempvar)
            tempvar = ''
    coords1.append(templist)
for triplets in coords1:
    x_val.append(float(triplets[0]))
    y_val.append(float(triplets[1]))
    z_val.append(float(triplets[2]))
    
###adjust start and end  

start = 0
end = 11

###

x_val2 = x_val[start:end]
y_val2 = y_val[start:end]
z_val2 = z_val[start:end]  

x_avg1 = []
y_avg1 = []
z_avg1 = []
for i in range(len(x_val2)):
    if i != len(x_val2)-2:
        x_a = st.mean(x_val2[i:i+3])
        y_a = st.mean(y_val2[i:i+3])
        z_a = st.mean(z_val2[i:i+3])
        x_avg1.append(x_a)
        y_avg1.append(y_a)
        z_avg1.append(z_a)
    else:
        x_a = st.mean(x_val2[i:])
        y_a = st.mean(y_val2[i:])
        z_a = st.mean(z_val2[i:])
        x_avg1.append(x_a)
        y_avg1.append(y_a)
        z_avg1.append(z_a)
        break

        
fig = plt.figure(figsize=(12,10))

ax = fig.add_subplot(111, projection='3d')

x_avg1 = [x_val2[0]] + x_avg1[:-1] + [x_val2[-1]]
y_avg1 = [y_val2[0]] + y_avg1[:-1] + [y_val2[-1]]
z_avg1 = [z_val2[0]] + z_avg1[:-1] + [z_val2[-1]]


ax.plot(x_val2,y_val2,z_val2)
ax.plot(x_avg1,y_avg1,z_avg1)
plt.show()
```
